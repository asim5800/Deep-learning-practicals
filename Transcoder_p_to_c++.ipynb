```python
# Star raring question coderbyte
def star_rating ( str ) :
    float(str)
    convert = float
    value = convert ( str )
    result = ""
    star_count = 0
    while star_count < 5 :
        if value >= 1 :
            result += "full "
            value -= 1
        elif value > 0 :
            if value +.25 >= 1 :
                result += "full "
            elif value +.25 >=.5 :
                result += "half "
            else :
                result += "empty "
            value -= value
        else :
            result += "empty "
        star_count += 1
    return result


print(star_rating ('2.36'))
```
